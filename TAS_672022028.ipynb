**SOAL 1**. Apa perbedaan Supervised Learning dan Unsupervised Learning? (Point 5)


Supervised Learning dan Unsupervised Learning menonjolkan perbedaan mendasar yang bergantung pada ketersediaan label dalam data yang digunakan untuk melatih model. Supervised Learning memerlukan data berlabel untuk memandu algoritma dalam memetakan input ke output yang diinginkan, seolah-olah mengandalkan peta untuk mencapai tujuan. Sebaliknya, Unsupervised Learning bekerja dengan data yang tidak memiliki label, menantang algoritma untuk menjelajahi dan mengungkap struktur atau pola tersembunyi tanpa petunjuk eksplisit, mirip dengan penjelajah yang menemukan jalan di hutan belantara. Dalam skenario ini, algoritma mengelompokkan data menjadi 'k' cluster berdasarkan kesamaan yang terdeteksi, dengan setiap cluster menampilkan karakteristik unik yang berbeda, menunjukkan kemampuan algoritma dalam menemukan tatanan dalam kekacauan yang tampak acak.

**c. Clustering berdasarkan gaji**

```python
#import library untuk melakukan penskalaan data
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Menghapus Data yang Missing
dataset_TAS.dropna(subset=['Salary'], inplace=True)

# Menyiapkan Data Gaji untuk Clustering
gaji = dataset_TAS['Salary'].values.reshape(-1, 1)

# Penskalaan Data
skala = StandardScaler()
gaji_skala = skala.fit_transform(gaji)

#Membuat dan Melatih Model K-Means
k = 4
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(gaji_skala)

# Menambahkan Label Cluster ke Dataset
dataset_TAS['Cluster'] = kmeans.labels_

# Mengatur ukuran gambar
plt.figure(figsize=(12, 4))
# Visualisasi data
plt.scatter(dataset_TAS['Age in Yrs.'], dataset_TAS['Salary'], c=dataset_TAS['Cluster'], cmap='magma', edgecolor='k', s=26)
# Menambah nama judul dan label sumbu
plt.title('Clustering berdasarkan Gaji')
plt.xlabel('Umur dalam tahun')
plt.ylabel('Gaji')
# Menampilkan data
plt.show()
```

**d. Pendapatan Bulanan berdasarkan lama bekerja di perusahaan.**

```python
#import libreries
from sklearn.linear_model import LinearRegression
import numpy as np

salary_column = 'Salary'
tenure_column = 'Age in Company (Years)'

# mengecek kolom yang kosong
if salary_column in dataset_TAS.columns and tenure_column in dataset_TAS.columns:
    # Menghapus Baris dengan Data Missing (NaN)
    dataset_TAS.dropna(subset=[salary_column, tenure_column], inplace=True)

    # Menyiapkan Data
    gaji_bulanan = dataset_TAS[salary_column]
    masa_jabatan = dataset_TAS[tenure_column]

    # Mengatur ukuran gambar
    plt.figure(figsize=(12, 6))
    plt.scatter(masa_jabatan, gaji_bulanan, color='deeppink', alpha=0.6, edgecolor='black')

    # Menambah nama judul dan label sumbu
    plt.title('Pendapatan Bulanan berdasarkan Lama Bekerja di Perusahaan')
    plt.xlabel('Lama Bekerja dalam tahun')
    plt.ylabel('Pendapatan per Bulan')

    # Persiapan Data untuk Regresi Linear
    t = masa_jabatan.values.reshape(-1, 1)
    i = gaji_bulanan.values.reshape(-1, 1)

    # Membuat dan Melatih Model Regresi Linear
    model = LinearRegression()
    model.fit(t, i)

    # Prediksi dan Visualisasi Garis Regresi
    t_tes = np.linspace(np.min(masa_jabatan), np.max(masa_jabatan), 100).reshape(-1, 1)
    i_prediksi = model.predict(t_tes)

    # Menambahkan Legenda dan Menampilkan Plot
    plt.plot(t_tes, i_prediksi, color='blue', linewidth=2.5, label='Linear Regression')

    # Untuk menampilkan legenda
    plt.legend()

    # Untuk menampilkan
    plt.show()
else:
    print("Dataset does not contain the required columns.")
```
