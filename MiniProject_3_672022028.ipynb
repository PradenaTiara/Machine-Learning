```python
#Import library
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
import matplotlib.ticker

from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
df = pd.read_csv('apel_orange.csv')
df.head()
```

```python
df['Class'].replace(['orange', 'apple'],
                        [0, 1], inplace=True)
```

```python
Weight = df['Weight'].values
Size = df['Size'].values
Class = df['Class'].values

scatter = plt.scatter(Weight[:50], Size[:50], c=Class[:50], cmap='cividis')
cbar = scatter
plt.xlabel('Weight')
plt.ylabel('Size')

##plt.legend(handles=scatter.legend_elements()[0], labels=['1. ', '2. E-Service', '3. Plus Service'], loc='upper left')

plt.show()
```

```python
x = df[['Weight', 'Size']].values
x[0:5]
```

```python
y = df[['Class']].values
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=4)
print('data training sebanyak: ', x_train.shape, y_train.shape)
print('data testing sebanyak : ', x_test.shape, y_test.shape)
```

```python
k = 12
neigh = KNeighborsClassifier(n_neighbors = k).fit(x_train, y_train)
neigh
```

```python
yhat = neigh.predict(x_test)
yhat[0:50]
```

```python
print("akurasi set train: ", metrics.accuracy_score(y_train, neigh.predict(x_train)))
print("akurasi set test: ", metrics.accuracy_score(y_test, yhat))
```
